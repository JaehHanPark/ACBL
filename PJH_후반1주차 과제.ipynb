```python
import pandas as pd

pileup = pd.read_csv('CLIP-let7g-gene.pileup', sep='\t', names=['chrom', 'pos', '_ref', 'count', 'basereads', 'quals'])
pileup.tail()
```

```python
from collections import Counter

def count_bases(s):
    counts = Counter(base.upper() for base in s if base.upper() in 'ATCG')
    total = sum(counts.values())
    return counts, total

pileup['base_counts'] = pileup['matches'].apply(lambda x: count_bases(x)[0])
pileup['total'] = pileup['matches'].apply(lambda x: count_bases(x)[1])
pileup[['pos', 'base_counts', 'total']].head()
```

```python
import numpy as np

def shannon_entropy(base_count):
    total = sum(base_count.values())
    if total == 0:
        return 0.0
    entropy = -sum((count / total) * np.log2(count / total) for count in base_count.values() if count > 0)
    return entropy

pileup['entropy'] = pileup['base_counts'].apply(shannon_entropy)
pileup[['pos', 'entropy']].head()
```

```python
import pandas as pd
import re
import numpy as np
from collections import Counter

# 파일 불러오기
pileup_d = pd.read_csv('CLIP-let7d-gene.pileup', sep='\t',
                       names=['chrom', 'pos', '_ref', 'count', 'basereads', 'quals'])

# 특수 문자 제거
toremove = re.compile('[<>$*#^]')
pileup_d['matches'] = pileup_d['basereads'].apply(lambda x: toremove.sub('', x))

# 염기 수 및 Shannon entropy 계산
def count_bases(s):
    counts = Counter(base.upper() for base in s if base.upper() in 'ATCG')
    return counts

def shannon_entropy(base_counts):
    total = sum(base_counts.values())
    if total == 0:
        return 0.0
    return -sum((count / total) * np.log2(count / total)
                for count in base_counts.values() if count > 0)

pileup_d['base_counts'] = pileup_d['matches'].apply(count_bases)
pileup_d['entropy'] = pileup_d['base_counts'].apply(shannon_entropy)
```

```python
output_file = 'let7d_entropy.bedgraph'

with open(output_file, 'w') as f:
    f.write('track type=bedGraph name="Mirlet7d_entropy" description="Shannon entropy of Mirlet7d" visibility=full color=200,0,0\n')
    for _, row in pileup_d.iterrows():
        start = int(row['pos']) - 1  # bedGraph는 0-based
        end = int(row['pos'])
        f.write(f"chr13\t{start}\t{end}\t{row['entropy']:.4f}\n")
```
