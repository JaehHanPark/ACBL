```python
!grep '	start_codon	.*	+	.*transcript_support_level "1"' ../binfo1-datapack1/gencode.gtf | \
 sed -e 's/	[^	]*transcript_id "\([^"]*\)".*$/	\1/g' > gencode-start.gtf
!head gencode-start.gtf
```

```python
# 아래 공백 중 일부는 "탭(Tab)" 문자입니다. 입력하기 힘들 때는 텍스트 편집기에서 입력해서 복사&붙여넣기 하면 됩니다.
!grep '	exon	.*	+	' ../binfo1-datapack1/gencode.gtf | \
 sed -e 's/	[^	]*transcript_id "\([^"]*\)".*$/	\1/g' > gencode-plusexon.gtf
!head gencode-plusexon.gtf
```

```python
# 아래 공백 중 일부는 "탭(Tab)" 문자입니다. 입력하기 힘들 때는 텍스트 편집기에서 입력해서 복사&붙여넣기 하면 됩니다.
!bedtools intersect -a gencode-start.gtf -b gencode-plusexon.gtf -wa -wb | \
 awk -F'	' -v OFS='	' '$9 == $18 { print $10, $13-1, $14, $18, $4-1, $16; }' | \
 sort -k1,1 -k2,3n -k4,4 > gencode-exons-containing-startcodon.bed
!head gencode-exons-containing-startcodon.bed; tail gencode-exons-containing-startcodon.bed
```

```python
!(samtools view -H ../binfo1-datapack1/RPF-siLuc.bam; \
  samtools view -F20 ../binfo1-datapack1/RPF-siLuc.bam | \
  bioawk -c sam '{ if (length($seq) >= 25) print $0; }') | \
 samtools view -b -o filtered-RPF-siLuc.bam
```

```python
!bedtools genomecov -ibam filtered-RPF-siLuc.bam -bg -5 > fivepcounts-RPF-siLuc.bed
!head fivepcounts-RPF-siLuc.bed
```

```python
!bedtools intersect -a fivepcounts-RPF-siLuc.bed -b gencode-exons-containing-startcodon.bed \
 -wa -wb -nonamecheck > fivepcounts-filtered-RPF-siLuc.txt
!head fivepcounts-filtered-RPF-siLuc.txt
```

```python
# 여기에 새로 코드를 작성해서 논문 그림과 비슷하게 만들어 봅시다.
import pandas as pd
import matplotlib.pyplot as plt

# 파일 경로
input_file = "/content/drive/MyDrive/binfo1-work/fivepcounts-filtered-RPF-siLuc.txt"

# 파일 읽기
df = pd.read_csv(input_file, sep='\t', header=None)

# 컬럼 이름 지정
df.columns = ['chr1', 'start1', 'end1', 'count', 'chr2', 'start2', 'end2',
              'transcript_id', 'start_codon', 'strand']

# 상대 위치 계산 (start codon 기준)
df['rel_pos'] = df['start1'] - df['start_codon']

# 원하는 범위 제한 (-50 ~ +50 nt)
df = df[(df['rel_pos'] >= -50) & (df['rel_pos'] <= 50)]

# 위치별 count 합산
summary = df.groupby('rel_pos')['count'].sum().reset_index()

# 시각화
plt.figure(figsize=(10, 5))
plt.bar(summary['rel_pos'], summary['count'] / 1000, width=1, color='black')

# 수직선 (start codon = 0)
plt.axvline(0, color='red')

# 시각적 강조 (start codon 주변 0 ~ +15)
plt.axvspan(0, 15, color='orange', alpha=0.2)

plt.title("Ribosome footprint near start codon (siLuc)")
plt.xlabel("Relative position to start codon of 5'-end of reads")
plt.ylabel("Raw read count (×1000)")
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```
