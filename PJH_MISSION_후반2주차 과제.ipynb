```python
import pandas as pd
cnts = pd.read_csv('read-counts.txt', sep='\t', comment='#', index_col=0)
cnts.head()
```

```python
cnts['clip_enrichment'] = cnts['CLIP-35L33G.bam'] / cnts['RNA-control.bam']
cnts['rden_change'] = (cnts['RPF-siLin28a.bam'] / cnts['RNA-siLin28a.bam']) / (cnts['RPF-siLuc.bam'] / cnts['RNA-siLuc.bam'])
cnts.head()
```

```python
from matplotlib import pyplot as plt
import numpy as np

fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.scatter(np.log2(cnts['clip_enrichment']),
           np.log2(cnts['rden_change']))
```

```python
from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import pearsonr

mask = (cnts['clip_enrichment'] > 0) & (cnts['rden_change'] > 0)
x = np.log2(cnts.loc[mask, 'clip_enrichment'])
y = np.log2(cnts.loc[mask, 'rden_change'])

r, _ = pearsonr(x, y)

fig, ax = plt.subplots(figsize=(4, 4))
ax.scatter(x, y, color='black', alpha=0.3, s=5)


ax.set_xlim(-10, 10)
ax.set_ylim(-8, 8)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.tick_params(direction='out', length=3)

ax.text(2.3, -1.8, f'$r$ = {r:.4f}', fontsize=10, style='italic')

plt.tight_layout()
plt.show()
```

```python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
mouselocal = pd.read_csv('https://hyeshik.qbio.io/binfo/mouselocalization-20210507.txt', sep='\t')
mouselocal.head()
```

```python
print(mouselocal.columns)
```

```python
# gene_id 중 마지막 것만 남기고 앞에 있는 중복 gene_id 컬럼 삭제
cnts = cnts.loc[:, ~cnts.columns.duplicated(keep='last')]
```

```python
print(mouselocal.columns)
```

```python
print(cnts.columns)
```

```python
# 인덱스 복원
cnts.reset_index(inplace=True)

# gene_id 컬럼으로 이름 변경
cnts.rename(columns={'Geneid': 'gene_id'}, inplace=True)

# log2 연산 필터
mask = (cnts['clip_enrichment'] > 0) & (cnts['rden_change'] > 0)
cnts = cnts.loc[mask].copy()
cnts['log2_clip'] = np.log2(cnts['clip_enrichment'])
cnts['log2_rden'] = np.log2(cnts['rden_change'])

# 병합
merged = cnts.merge(mouselocal[['gene_id', 'type']], on='gene_id', how='inner')
```

```python
# 버전 제거된 gene_id인지 먼저 확인
cnts['gene_id'] = cnts['gene_id'].str.split('.').str[0]

# 겹치는 ID 수
overlap = set(cnts['gene_id']).intersection(set(mouselocal['gene_id']))
print(f'겹치는 gene_id 개수: {len(overlap)}')
```

```python
print(f"필터링 후 cnts 크기: {cnts.shape[0]}")
```

```python
merged = cnts.merge(mouselocal[['gene_id', 'type']], on='gene_id', how='inner')
print(f"병합 후 merged 크기: {merged.shape[0]}")
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# 상관계수 계산
r, _ = pearsonr(merged['log2_clip'], merged['log2_rden'])

# 고유 type 리스트와 색상 매핑
types = merged['type'].unique()
colors = plt.cm.tab10.colors
color_map = dict(zip(types, colors[:len(types)]))

# plot 그리기
fig, ax = plt.subplots(figsize=(6, 5))

for t in types:
    sub = merged[merged['type'] == t]
    ax.scatter(sub['log2_clip'], sub['log2_rden'],
               color=color_map[t], label=t, s=10, alpha=0.6)

ax.set_xlim(-10, 10)
ax.set_ylim(-8, 8)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.tick_params(direction='out', length=3)
ax.set_xlabel('log2(CLIP enrichment)')
ax.set_ylabel('log2(RDEN change)')
ax.text(2.3, -1.8, f'$r$ = {r:.4f}', fontsize=10, style='italic')

ax.legend(title='type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
from gprofiler import GProfiler

# gProfiler 객체 생성
gp = GProfiler(return_dataframe=True)

# gene ID 리스트 준비
gene_ids = merged['gene_id'].unique().tolist()

# GO term enrichment 수행 (생쥐 기준)
go_results = gp.profile(organism='mmusculus', query=gene_ids)

# 필요한 열만 정리
go_results = go_results[['native', 'name', 'p_value', 'intersection_size']]
go_results['log10_FDR'] = -np.log10(go_results['p_value'])
```

```python
import numpy as np

# 시드 고정으로 재현 가능하게
np.random.seed(42)

# X, Y 위치 무작위로 생성
go_results['x'] = np.random.uniform(-2, 4, size=len(go_results))
go_results['y'] = np.random.uniform(-1, 1.5, size=len(go_results))

# 복사해서 최종 시각화용 테이블 생성
go_plot = go_results.copy()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 6))

# 점의 크기 = gene 개수
sizes = go_plot['intersection_size'] * 5

# 점의 색상 = -log10 FDR
colors = go_plot['log10_FDR']

# Bubble Plot
sc = ax.scatter(go_plot['x'], go_plot['y'],
                s=sizes,
                c=colors,
                cmap='YlOrRd_r',
                edgecolors='black',
                alpha=0.7)

# FDR 가장 낮은 10개 라벨 표시
top_terms = go_plot.nsmallest(10, 'p_value')
for _, row in top_terms.iterrows():
    ax.text(row['x'], row['y'], row['name'], fontsize=8, ha='center')

# 축 및 스타일
ax.axhline(0, color='gray', linestyle='--')
ax.axvline(0, color='gray', linestyle='--')
ax.set_xlabel('Enrichment level of LIN28A-bound CLIP tags (log2) — random')
ax.set_ylabel('Ribosome density change upon Lin28a knockdown (log2) — random')
ax.set_title('GO Term Enrichment Bubble Chart (Randomized Coordinates)')

# 컬러바 추가
cbar = plt.colorbar(sc, ax=ax)
cbar.set_label('Term-specific enrichment confidence (-log10 FDR)')

plt.tight_layout()
plt.show()
```
